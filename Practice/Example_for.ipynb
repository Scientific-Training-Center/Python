```python
import matplotlib.pyplot as plt

# Define the range of x values
x_values = range(-10, 11)

# Calculate the y values for each x value using a for loop
y_values = []
for x in x_values:
    y = x**2
    y_values.append(y)

# Plot the graph
plt.plot(x_values, y_values)

# Add labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# Display the graph
plt.show()
```
